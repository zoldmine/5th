```python
import datetime
import numpy as np
import scipy as sp
import scipy.fftpack
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('weather.csv', na_values=(-9999),parse_dates=['DATE'])
data
```

```python
df = data.dropna().groupby('DATE').mean()
df
```

```python
temp = (df['TMAX']+df['TMIN'])/20
```

```python
temp
```

```python
date = df.index
```

```python
N = len(temp)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6)) 
temp.plot(ax=ax, lw=.5) # 
ax.set_ylim(-10, 40) 
ax.set_xlabel('Дата')
ax.set_ylabel('Средняя температура')
```

```python
temp_fft = np.fft.fft(temp)
```

```python
temp_psd = np.abs(temp_fft) ** 2
```

```python
fftfreq = sp.fftpack.fftfreq(len(temp_psd), 1./365)
```

```python
plt.plot(fftfreq)   
```

```python
i = fftfreq > 0
plt.plot(i)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6)) 
ax.plot(fftfreq[i], 10 * np.log10(temp_psd[i])) 
ax.set_xlim(0, 5)
ax.set_xlabel('Частота (1/год)')
ax.set_ylabel('Спектральная плотность мощности (дБ)')
```

```python
temp_fft_bis = temp_fft.copy() 
temp_fft_bis[np.abs(fftfreq) > 1.2] = 0 
```

```python
temp_slow = np.real(sp.fftpack.ifft(temp_fft_bis))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6)) 
temp.plot(ax=ax, lw=.5)                       
ax.plot_date(date, temp_slow, '-', lw=2)            
ax.set_xlim(datetime.date(1994, 1, 1),
            datetime.date(2014, 1, 1))
ax.set_ylim(-10, 40)
ax.set_xlabel('Даты')
ax.set_ylabel('Средняя температура')
```
